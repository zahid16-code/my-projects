## Droping the duplicate values

```python
#dropping duplicate values with # columns
df=df.drop_duplicates(subset=["#"], keep='first')
df
```

```python
#duplicate values has been deleted
df.head(10)
```

```python
#showing the unique values after cleaning duplicate values
df.nunique()
```

## Filling the null values

```python
#null values in the data frame
df.isnull()
```

```python
#number of null values in the data frame
df.isnull().sum()
```

```python
#filling the null values 
df.update(df['Type 2'].fillna('Unknown'))
```

```python
#after filling the null values
df.head(15)
```

```python
#checking null values
df.isnull().sum()
```

```python
# deleting the "#" column
df=df.drop("#",axis=1)
df.head(10)

```

Data cleaning of the given data is completed!

```python
# cleaned data storing in csv file
df.to_csv('output.csv', index=False)
```

### Data visualisations

Data visualisations can be done with by taking meaning insights and visualise it

```python
import matplotlib.pyplot as plt
```

```python
#classification of types of pokemon
types=df["Type 1"].unique()
type_counts = df['Type 1'].value_counts()
types
```

```python
#visualisations of classified pokemons
plt.pie(type_counts,labels=types)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.show()
```
